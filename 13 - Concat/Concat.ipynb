```python
import pandas as pd
df1 = pd.read_csv('bank.csv', encoding='utf-8', sep=';')
df2 = pd.read_csv('Dados aula 1.csv', encoding='utf-8', sep=';')
```

```python
#Concatena duas dataframes
pd.concat([df1,df2]).head()
```

```python
#Limpa e reseta os indices
pd.concat([df1,df2], ignore_index=True)
```

```python
#Adiciona um index hierarquico
pd.concat([df1,df2], keys=['primeiro','segundo'])
```

```python
#names -> Nomeia os indices na hierarquia
pd.concat([df1,df2], keys=['primeiro','segundo'], names=['Tiranu','pao'])
```
